**This notebook will reload (Truncate and load) all the tables from the csv files in the AutoClaims_csv folder in the lakehouse**

```python
table_list = ["accident","adjuster","claim","policy","policyholder","vehicle"]
```

```python
# Loop through all configured tables and execute the load function
for table in table_list:

    #Get schema of table
    existing_df = spark.read.table(table)
    target_schema = existing_df.schema

    #Read the csv file
    df = spark.read \
        .format("csv") \
        .option("header","true") \
        .schema(target_schema) \
        .load("Files/AutoClaims_csv/"+table+".csv")

    #overwrite into the table
    df.write.mode("overwrite").saveAsTable(table)

    print(f"Table {table} loaded")
```
